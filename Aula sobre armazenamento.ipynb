```python
# para colocar a bibliotecaa em uso, devemos importar ela
# importe NOME_DA_BIBLIOTECA as APELIDO_PARA_USO
import pandas as pd
```

```python
#criar o 'esqueleto' do arquivo

# vammos usar dicionarios para criar os dados
# dicionario funciona com a ideia de chave-valor ou chave-valores

dados = {
    'nome':['Katia','Douglas','Matheus','Thiago'],
    'idade':[32,28,23,20],
    'cidade':['Campinas','Campinas','Campinas','Campinas']
}

# a pandas pode transformar um dicionario em uma 'tabela'
# esse pedaço de tabela é chamado de DataFrame

# para transformar
df_dados = pd.DataFrame(dados)

display(df_dados)
```

```python
#criar um arquivo csv ('excel')

# para criar, precisamos de dados
# chamamos os dados e atribuimos a função to_csv/doc/ect

df_dados.to_csv('teste_dados.csv', index=False)
```
