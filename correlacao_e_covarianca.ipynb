```python
import numpy as np

alturas = np.array([160, 170, 175, 163, 180])
# peak-to-peak
print(alturas)
amplitude = np.ptp(alturas)
print(amplitude)
# amplitude = maior valor - menos valor
```

```python
# desvio padrão
media_alturas = np.mean(alturas)
print(media_alturas)

#calculo de desvio padrão
desvio_padrao = np.std(alturas)
print(desvio_padrao)
```

```python
#covariancia positivo
# duas coisas crescem ou decrescem ao mesmo tempo

#covariancia negativa
#quando uma variavel cesce e a outra decresce

#covariancia zero
#nada muda

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

dados_equipamento = {
    'consumo' : [10,20,30,40,50,60,70,80],
    'temperatura' : [25,32,45,48,60,72,85,90]
}

#transformar esses dados para o pandas ler
df_dados = pd.DataFrame(dados_equipamento)

# vamos fazer o calculo de covariancia
# criar uma base para a covariancia
cov_matrix = df_dados.cov()
print(f'a covariancia foi {cov_matrix.iloc[0,1]}')

# corr - função para correlação
correlacao_matrix = df_dados.corr()
print(correlacao_matrix)

# gerar um grafico para essas correlações
sns.scatterplot(x='consumo', y='temperatura', data=df_dados)
plt.title('grafico de correlação')
plt.show()
```
