```python
import requests
import json
#paara coletar data e hora
from datetime import datetime

#vamos colocar nossaa endpoint
url = 'https://script.google.com/macros/s/AKfycbzYp29a67ZZQeG1texHALuA9Zcplz5DHALI7Zg5JuYsGL6BbDUH1wuPY4Jqe6X2ZWNu/exec'

# agora vamos criar a função que envia os dados para o excel
def enviar_dados(placa, carro, funcionario):
  # vou colocaar aqui a data e hora
  data_e_hora = datetime.now().strftime('%d/%m/%y %H:%M:%S')
  #%d - dia
  #%m - mes
  #%y - ano
  #%H - hora
  #%M - minuto
  #%S - segundo
  #datetime.now() - salva o momento que o codigo é executado
  parametros = {
      'data_hora': data_e_hora,
      'placa' :placa,
      'carro' :modelo,
      'funcionario' :funcionario
  }
  print('enviando dados para a tabela')

  resposta = requests.post(url, json=parametros)

  if resposta.status_code == 200:
    print('Dados proontos para envvio')
  else:
    print(f'erro ao enviar os dados: {resposta.status_code}')
```
